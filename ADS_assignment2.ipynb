```python
employees.groupby(['JobRole', 'Attrition', 'DistanceFromHome']).count()
```

```python
resigned_employees = employees[employees['Attrition'] == 'Yes']
```

```python
total_resigned = resigned_employees["Age"].count()
#lets now calculate the percentage of employees who resigned to total workforce
total_resigned/employees['Age'].count() * 100
```

```python
#so 16% of the workforce resigned
```

```python
#What is the gender balance on resignations like?
resigned_employees['Gender'].value_counts()
```

```python
#The percentages per gender on resignations
females_resigned = resigned_employees[resigned_employees['Gender'] == "Female"]
males_resigned = resigned_employees[resigned_employees['Gender'] == "Male"]

noMale = males_resigned['Gender'].count()
noFemale = females_resigned['Gender'].count()
noFemale/(noMale + noFemale)*100
```

```python
#what was the percentage of females in the whole workforce?
employees[employees['Gender']=='Female'].count()['Age']/(employees[employees['Gender']=='Female'].count()['Age'] +
                                                        employees[employees['Gender']=='Male'].count()['Age'])*100
```

```python
#So women made 40% of the entire workforce before attrition and they made 36% of resignations therefore more males are resigning
```
